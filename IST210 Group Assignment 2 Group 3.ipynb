Dishes
	Dishes can all be found and defined by name, which serves as the primary key. This also acts as the foreign key in transactions for price retrieval. Each dish has a price associated with it. In addition, the number of times it is ordered is tracked to allow the rotation of underperforming dishes to occur.  The ingredients serve as a list to retrieve how much of each ingredient is in stock. Using these values the Number in stock is calculated to determine how much of the dish can be made before running out of the required ingredients.

-Name
-numberSold
-Price (double > 0)
-Ingredients
-NumberInStock (int > 0)

Shipments

-Item Type

Employees

For the entity of employees the primary key is the EmployeeID. With this EmployeeID each employee is given a unique string of 5 numbers that can be used to identify each individual employee. The next following attributes (FirstName, LastName, Address, and PhoneNumber) provide more identifying information about the employees that the restaurant may need. The phone number attribute allows the restaurant to collect the phone numbers of all of their employees and provides a way for the restaurant to get in contact with the employees if need be. The attribute of position designates what job the employee was hired to work in the restaurant. The EmployeeID primary key is also related to the CustormerID primary key of the customer entity. The purpose of this relationship is to keep a log of which and how many  employees interact with customers. This relationship can range from 0 to 5 employees interacting with up to 100 customers. The employee entity is also related to the transaction entity. The EmployeeID is related in a one to one relationship to the TransactionID and this is done to keep track of each transaction an employee processes.

-EmployeeID (5 digits)
-FirstName 
-LastName
-Address
-PhoneNumber (10 digits)
-Position


Transactions
The transaction entity has a TransactionID as a primary key so that each transaction can be stored and identified using a unique identity. Within the transaction entity there's a foregin key of a CustomerID from the customer  entity and this is used to relate a transaction to a customer. The primary key of TransactionID also as a relationship to the entities of dishes, rewards cards and employees too. As previously stated the TransactionID and EmployeeID are related to link an employee to a transaction. The relationship between TransactionID and dishes or rewards cards works the same way in that it is used to relate a unique entity to the transaction. The transaction entity also has the attributes of Datetime which tracks the time that a transaction is made. The PaymentType attribute is responsible for recording how the customer paid for their transaction. 

-TransactionID (10 digits)
-Datetime
-CustomerID	
-PaymentType
-Reward Discount (double < 100%) 

Rewards card
The rewards cards can be identified with the 10 digit card ID, which serves as our primary key for this entity. The customers first and last name are also key attributes for this entity and they are crucial because the rewards card has to correctly and fully identify each person that is a member. The customer's phone number allows the restaurant to text the person info about their rewards points, special deals, waitlist information, etc. The address is also included to not only further describe each member in the system, but also if the restaurant needs to mail any information about their rewards points or special offers. Lastly, loyalty points is included and is a necessity here because it is how we keep track of how “loyal” of a customer they are meaning how much money they are spending at the restaurant, which is then how we determine what special deals we offer them. 

CardID (10 digits)
-FirstName
-LastName
-Phone (10 digits)
-Address
-Loyalty points

Customers
For the customers entity we are identifying and tracking the information necessary to characterize each individual customer. Every customer is primarily and most easily classified by their first name, which is why it is our primary key, and last name is also included to make sure their name is as descriptive and specific as possible in the system. Phone number is included because it is the way the restaurant can identify their customers most easily without having to ask for their name, which comes in handy especially for people with longer names and when taking orders over the phone. Address is crucial for knowing where to deliver to customers. Lastly, rewards card is included here because it is the way we can tell which customers are repeat customers or first-timers.

-LastName
-Phone (10 digits)
-Address
-Rewards card
