In our ER diagram we chose to include 9 entity tables including: 

Customer
In terms of the customers attributes we need to know their first and last name in order to identify them on a personal level and a customerID to identify them in a numerical medium, such as our database. As secondary keys we have the customers address and phone number in case we would need to differentiate between customers or look up a customer. Also the rewards card is tied to these other attributes in this data table to ensure each customer has his own rewards card ID.

rewardCard
This entity has two attributes, it has the customer ID and the customers accompanying point value specific to that customer.

animalPurchases
For this entity, we are capturing data about who bought the animals (be it fish, pets, or both) and how many of each they purchased. Within the transaction, if the customer has a reward card, they may be eligible for a discount. Some information about the customer and which animal they buy is logged for future reference. For store use, were logging which specific animals they buy, so we can keep our system (health records for individual animals, adoption papers, etc.) up to date.

Food + foodPurchases
For the food entity the major relationship for the entity is it being related to the foodPurchase entity. The primary key of the food entity is a unique barcode to act as the primary key of the entity and it keeps track of each individual unit of food. The secondary key of food are name and the price. This help give distinguishing features to the different units of food. This barcode primary key is then used in the foodPurchases entity as a foregin key to show a many to many relationship between food and foodPurchases. foodPurchases has a primary key of foodPurchaseid to give it a unique identity and then the secondary key of customerid is a foregin key from the customer entity to relate the two entities together.

Fish
For the fish entity table we are keeping track of all the fish that come through the pet store, the aquarium and aquarium type they reside in along with the particular species of fish. The primary key of the fish entity is the fishID, which is a 10 digit number that tracks every fish that has been part of the pet stores inventory. For our secondary keys we are tracking the aquarium ID, aquarium type, and species so that the location of each particular fish in the store can be easily be identified, making transactions as smooth as possible. 

Aquarium
For the aquarium entity table we are keeping track of the specifics behind every aquarium in the pet store. Our primary key for this table is the aquarium ID, which is a five digit number that makes it easy to store and update information in a database about all the aquariums. For our secondary keys we are tracking the aquarium type along with the maximum current capacity, which makes it much easier for the pet store employees to know where they can fit more fish, turtles, frogs, etc., keeping all that data organized and easily accessible. 

Pets
This table handles the inventory of pets and what enclosures they belong in. This allows for the easy retrieval of the pet when purchasing/adopting. Each pet has a unique ID so when searched for at checkout can be found easily. The enclosure number and breed combined can be used as a composite secondary key to retrieve the petID if it is unknown. The petID shows up as a foreign key on the transaction record when adopting.

Enclosure
The enclosures are relatively simple, assigned a unique number, a type (pen, cage etc.), maximum capacity and current capacity. The enclosure number acts as a foreign key under the pets table in order to locate a specific pet.
