Restaurant on the Water  This idea is unique as it introduces variables about food such as prep time, price, expiration dates, and other variables not found in standard products. This is a new idea because our customers have to access the drive-through window via boat. Typical business interactions include our suppliers dropping off fresh food from a barge in the mornings. All intake of ingredients will be logged and subtracted from the inventory when used. After that, we'll prepare the ingredients and open up shop. Customers will then drive up to the window and purchase food. We collect data about the transactions including the time of the transaction, item bought, price, etc. to help us predict what future inventory we need and trends in customer behavior. This will be used to see what dishes are being sold the most, which ones are rarely sold and should be swapped for a new dish. This will also help coordinate sales/deals on the most popular dishes that would encourage customers to come more frequently. For customers that are frequent visitors to the restaurant, a reward card will be offered for them to redeem rewards for purchases and loyalty helping them save money. Each customer will be given a 10 digit ID number and a points account with their name, phone number, address and number of points. The point system would be based off of how much was spent per purchase including tax. These offers may include standard and rotating rewards, an example of a standard reward would be a free side or an upgrade in drink size, rotating may be seasonal drinks of dishes.

Classroom/School  At a school, there are many different types of classrooms with unique needs. A database of supplies would provide a streamline of organizing all of these resources. For example it would catalog the IT active, stored or under repair devices or for a chem lab all the chemicals, glassware, burners and other such items being used, broken etc during the day. whenever a shipment of supplies come in they are cataloged according to what subject they are being used for (English, Science, History, Math Foreign Language, IT) and are then sent to a specified storage closet to be stored until use. When taken from the closet the items are checked out as returnable, one time use, permanent. returnable means its an item for IT such as a temporary laptop, a one time use would be like a pack of pencils, and permanent would be burners and glassware for a science lab. Each teacher and faculty have ID numbers attached to their checkouts, so all items are accounted for by who it was checked out and what classroom it's being taken to. Each month supplies used and left over are recorded and then orders are adjusted for the next month. Ideally each department has a scheduled week of deliveries so as not to overwhelm the system. This database could also be expanded to include the food supplies for vending machines and cafeterias. In addition it can take special requests that are outside of normal ordering parameters if there is a specific event, activity or lab that a teacher wants to do.

Car Rental
For this idea we would be tracking the interactions at a Car Rental Company including the cars, store locations, customers (dynamic), transactions, and tires. We decided to choose a rental car company because they are very dynamic businesses that have many different components ranging from the actual cars and their elements to the customers and what makes up the transactions. A database is extremely necessary with a business model like this as you would need it to store and track all these different entities, especially because we are looking at a national car rental company from a corporation point of view, meaning that all the individual store locations will be subsets of what we are tracking. What makes a car rental company unique from other businesses and enterprises that require databases is the fact that a car rental company deals with actually loaning their inventory out. Also, car rental companies need to be even more advanced in keeping track of their entities since their whole business model consists of cars being returned at different locations, and moving inventory around to different locations depending on the demand and supply at that time. This is not a brand new business, as there are many existing car rental companies including Enterprise, Hertz, Thrifty, etc., but where we differentiate ourselves is we rent to anyone that has a license, whether they’re 16 and from Kansa or 93 and from Lithuania. Now this does provide some insurance problems but that is okay as our car rental company also services vehicles of all makes and models and we do not discriminate based on an old year or high mileage like all other car rental companies. By being able to significantly lower the cost of our cars, we’re able to balance out the insurance costs of renting to younger and older drivers. Also, because of our unique car rental methods, we are pretty much the only car rental company providing an option for marginalized groups in the community, giving car rental opporutnies to all peoples without discrimination, giving us a leg up on competitors. A typical business interaction at the car rental company would consist of a person, who could be a returning customer with a membership account or a first time customer. This customer could either book online or in person a particular car for a designated number of days. That customer, particularly if they are new, will have to give the company a lot of personal information, including their full name, address, credit card info, social security, insurance, etc. The reason being is because loaning out a car that’s worth anywhere from 10 to 60 thousand dollars is a big risk so it is important for the car rental company to protect themselves from the start. By making customers sign different contracts and agreements and by having their financial and personal information on file, the company is very well-prepared if something goes wrong. All in all, car rental companies are very unique businesses that require a different approach to inventory and transaction tracking due to their lending nature. It is important to keep in mind that the backbone of every successful business, whether that be a car rental company or a fast food restaurant, has an acute and efficient method of database storage to keep track of all their entities.
